```python
from enum import StrEnum


class Microservice(StrEnum):
    AUTH = "AUTH"
    EVENTS = "EVENTS"
    POSTS = "POSTS"
    USERS = "USERS"
    MEDIA = "MEDIA"
    LIVESTREAM = "LIVESTREAM"
    R18E = "R18E"

    def needs_rmq(self) -> bool:
        return self in (Microservice.R18E, Microservice.MEDIA, Microservice.POSTS, Microservice.LIVESTREAM, Microservice.EVENTS)


x = Microservice("MEDIA")
print(x.needs_rmq())
```
