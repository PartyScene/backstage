```python
import timeit
import uuid
from fastuuid import uuid4 as fastuid4
import uuid_utils as ruuid
import ulid
import shortuuid
from nanoid import generate

COUNT = 100_0000

print("Benchmarking ID generators...")

# uuid4
uuid4_time = timeit.timeit(lambda: uuid.uuid4(), number=COUNT)
print(f"uuid.uuid4() took {uuid4_time:.4f}s for {COUNT} calls")


# FastUUID
fast_time = timeit.timeit(lambda: str(fastuid4()).split('-')[-1], number=COUNT)
print(f"Fast UID4() took {fast_time:.4f}s for {COUNT} calls")


# RustUUID
rust_time = timeit.timeit(lambda: str(ruuid.uuid4()).split('-')[-1], number=COUNT)
print(f"Rust.UUID4() took {rust_time:.4f}s for {COUNT} calls")


# uuid1
uuid1_time = timeit.timeit(lambda: uuid.uuid1(), number=COUNT)
print(f"uuid.uuid1() took {uuid1_time:.4f}s for {COUNT} calls")

import os

# ulid
ulid_time = timeit.timeit(lambda: str(ulid.from_randomness(os.urandom(10))), number=COUNT)
print(f"ulid.ULID() took {ulid_time:.4f}s for {COUNT} calls")

# shortuuid
shortuuid_time = timeit.timeit(lambda: shortuuid.uuid(), number=COUNT)
print(f"shortuuid.uuid() took {shortuuid_time:.4f}s for {COUNT} calls")

# nanoid (default size = 21)
nanoid_time = timeit.timeit(lambda: generate(), number=COUNT)
print(f"nanoid.generate() took {nanoid_time:.4f}s for {COUNT} calls")
```
